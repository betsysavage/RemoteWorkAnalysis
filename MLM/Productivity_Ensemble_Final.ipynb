```python
columns = [
"birth_year",
"gender",
"industry_desc",
"occupation_desc",
"organization_size",
"manage_others",
"household",
"years_at_job",
"metro_or_regional",
"rw_percentage_2020",
"org_encouraged_rw",
"org_prepared_for_rw",
"rw_is_common_at_org",
"rw_permission_is_attainable",
"rw_collaboration_easy",
"preferred_rw_percentage_2020",
"preferred_rw_percentage_future",
"if_no_covid_employer_encourage_rw",
"if_no_covid_employer_support_rw",
"if_no_covid_i_would_have_choice_about_rw",
"productivity_remote_vs_office",
"inperson_hours__commuting",
"inperson_hours__working",
"inperson_hours__personal_family_time",
"inperson_hours_domestic_responsibilities",
"remote_hours_commuting",
"remote_hours_working",
"remote_hours_personal_family_time",
"remote_hours_domestic_responsibilities",
]

target = ["productivity_remote_vs_office"]
```

```python
# # Load the data
# file_path = Path('merged_df_cleaned_040323.csv')
# df = pd.read_csv(file_path)
df = merged_df.loc[:, columns].copy()

# Convert the target column values to More_or_same_productive and less_productive based on their values
x = dict.fromkeys(['more productive','much more productive'], 'more_productive')
df = df.replace(x)

x = dict.fromkeys(['same productivity'], 'same_productive')
df = df.replace(x)

x = dict.fromkeys(['less productive', 'much less productive'], 'less_productive')    
df = df.replace(x)

df.reset_index(inplace=True, drop=True)

df.head()
```

```python
## Split the Data into Training and Testing
```

```python
# Create our features
X = df.drop(columns= 'productivity_remote_vs_office')
X = pd.get_dummies(X)

# Create our target
y = df[target]
```

```python
X.describe()
```

```python
# Check the balance of our target values
y['productivity_remote_vs_office'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Resample the training data with the BalancedRandomForestClassifier
# Import dependencies and create a random forest classifier.
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(class_weight = "balanced", n_estimators=128, random_state=78) 

# Fitting the model
rf_model = rf_model.fit(X_train, y_train)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
y_pred = rf_model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score

balanced_accuracy_score(y_test, y_pred)
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```

```python
# List the features sorted in descending order by feature importance
importances = rf_model.feature_importances_

sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```

```python
# Tuning our random forest model for best results
# Used Harsh Kumar's RF tuning method: https://www.kaggle.com/code/harsh1kumar/santander-trxn-pred-rf-hp-tune/notebook
```

```python
#get the list of hyperparameters that can be tuned for a RandomForest Model
RandomForestClassifier().get_params()
```

```python
#Specify the range of values for some important hyperparameters
n_estimators_list = [130,160,190]
criterion_list = ['gini']
max_depth_list = [35,55]
min_samples_split_list = [0.001, 0.005]
min_samples_leaf_list = [0.001, 0.005]
max_features_list = ['log2']

params_grid = {
    'n_estimators': n_estimators_list,
    'criterion': criterion_list,
    'max_depth': max_depth_list,
    'min_samples_split': min_samples_split_list,
    'min_samples_leaf': min_samples_leaf_list,
    'max_features': max_features_list
}

num_combinations = 1
for k in params_grid.keys(): num_combinations *= len(params_grid[k])

print('Number of combinations = ', num_combinations)
params_grid
```

```python
#define area under the curve function and test all combinations of hyper parameters
def my_roc_auc_score(model, X, y): return metrics.roc_auc_score(y, model.predict(X))

rf_model_tuning = GridSearchCV(estimator=RandomForestClassifier(class_weight='balanced'),
                        param_grid=params_grid,
                        cv=3,
                        scoring=my_roc_auc_score,
                        return_train_score=True,
                        verbose=4)

rf_model_tuning.fit(X,y)
```

```python
# determine best possible parameters 
rf_model_tuning.best_params_
```

```python
#Use best parameters from previous step to optimally tune our model
rf_model_tuned = RandomForestClassifier(class_weight='balanced',
                                      criterion='gini',
                                      max_depth=35,
                                      max_features='log2',
                                      min_samples_leaf=0.001,
                                      min_samples_split=0.001,
                                      n_estimators=130)
rf_model_tuned.fit(X,y)
```

```python
# Display the tuned confusion matrix
y_pred = rf_model_tuned.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Calculated the tuned balanced accuracy score
balanced_accuracy_score(y_test, y_pred)
```

```python
# print the tuned imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```

```python
# List the features sorted in descending order by feature importance
importances_tuned = rf_model_tuned.feature_importances_

sorted(zip(importances_tuned, X.columns), reverse=True)
```
