```python
import pandas as pd
```

```python
# Read cleaned csv
cleaned_df = pd.read_csv('merged_df_cleaned_040323.csv', index_col= False)
cleaned_df.head()
```

```python
# create csv for respondents table
respondents_df = cleaned_df[['responder_id', 'birth_year', 'gender',
    'industry_desc', 'occupation_desc', 'manage_others',
    'household', 'years_at_job', 'metro_or_regional']].copy()
respondents_df.head()
```

```python
# Save respondents dataframe as csv file
respondents_df.to_csv("respondents.csv", index=False)
```

```python
# create csv for orginizational data table
orginizational_data_df = cleaned_df[['responder_id', 'organization_size', 'org_encouraged_rw', 
        'org_prepared_for_rw', 'rw_is_common_at_org', 'rw_permission_is_attainable', 
        'rw_collaboration_easy', 'if_no_covid_employer_encourage_rw',
        'if_no_covid_employer_support_rw', 'if_no_covid_i_would_have_choice_about_rw']].copy()
orginizational_data_df.head()
```

```python
# Save orginizational_data dataframe as csv file
orginizational_data_df.to_csv("orginizational_data.csv", index=False)
```

```python
# create csv for time management data table
time_management_df = cleaned_df[['responder_id', 'rw_percentage_2020', 
    'preferred_rw_percentage_2020', 'preferred_rw_percentage_future', 
    'productivity_remote_vs_office', 'inperson_hours__commuting', 
    'inperson_hours__working', 'inperson_hours__personal_family_time', 
    'inperson_hours_domestic_responsibilities','remote_hours_commuting',
    'remote_hours_working', 'remote_hours_personal_family_time',
    'remote_hours_domestic_responsibilities', 'commute_time_difference']].copy()
time_management_df.head()
```

```python
# Save time management dataframe as csv file
time_management_df.to_csv("time_management.csv", index=False)
```
